### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print(f"Faturamento total da Loja 1: R$ {faturamento_loja1:.2f}")
print(f"Faturamento total da Loja 2: R$ {faturamento_loja2:.2f}")
print(f"Faturamento total da Loja 3: R$ {faturamento_loja3:.2f}")
print(f"Faturamento total da Loja 4: R$ {faturamento_loja4:.2f}")
```

```python
import matplotlib.pyplot as plt

# Dados
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamento = [1534509.12, 1488459.06, 1464025.03, 1384497.58]

# Cores opcionais
cores = ['gold', 'lightcoral', 'lightblue', 'lightgreen']

# Gráfico de pizza
plt.figure(figsize=(8, 6))
plt.pie(
    faturamento,
    labels=lojas,
    autopct='%1.1f%%',
    startangle=90,
    colors=cores,
    shadow=True
)
plt.title('Participação no Faturamento Total por Loja')
plt.axis('equal')  # Mantém o círculo proporcional
plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
vendas_por_categoria_loja1 = loja['Categoria do Produto'].value_counts()
vendas_por_categoria_loja2 = loja2['Categoria do Produto'].value_counts()
vendas_por_categoria_loja3 = loja3['Categoria do Produto'].value_counts()
vendas_por_categoria_loja4 = loja4['Categoria do Produto'].value_counts()

print("Vendas por Categoria (Loja 1):")
display(vendas_por_categoria_loja1)

print("\nVendas por Categoria (Loja 2):")
display(vendas_por_categoria_loja2)

print("\nVendas por Categoria (Loja 3):")
display(vendas_por_categoria_loja3)

print("\nVendas por Categoria (Loja 4):")
display(vendas_por_categoria_loja4)
```

# 3. Média de Avaliação das Lojas

```python
media_avaliacao_loja1 = loja['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

print(f"Média de Avaliação da Loja 1: {media_avaliacao_loja1:.2f}")
print(f"Média de Avaliação da Loja 2: {media_avaliacao_loja2:.2f}")
print(f"Média de Avaliação da Loja 3: {media_avaliacao_loja3:.2f}")
print(f"Média de Avaliação da Loja 4: {media_avaliacao_loja4:.2f}")
```

```python
import matplotlib.pyplot as plt

# Dados das médias
medias = {
    'Loja 1': 3.98,
    'Loja 2': 4.04,
    'Loja 3': 4.05,
    'Loja 4': 4.00
}

# Criação do gráfico de barras
plt.figure(figsize=(8, 5))  # Tamanho do gráfico
plt.bar(medias.keys(), medias.values(), color='skyblue', edgecolor='black')
plt.title('Média de Avaliação por Loja')
plt.xlabel('Lojas')
plt.ylabel('Média da Avaliação')
plt.ylim(3.9, 4.1)  # Zoom nos valores para destacar as diferenças sutis

# Exibir os valores acima de cada barra
for loja, media in medias.items():
    plt.text(loja, media + 0.005, f'{media:.2f}', ha='center', va='bottom')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
vendas_por_produto_loja1 = loja['Produto'].value_counts()
vendas_por_produto_loja2 = loja2['Produto'].value_counts()
vendas_por_produto_loja3 = loja3['Produto'].value_counts()
vendas_por_produto_loja4 = loja4['Produto'].value_counts()

print("Produtos Mais Vendidos (Loja 1):")
display(vendas_por_produto_loja1.head())

print("\nProdutos Menos Vendidos (Loja 1):")
display(vendas_por_produto_loja1.tail())

print("\nProdutos Mais Vendidos (Loja 2):")
display(vendas_por_produto_loja2.head())

print("\nProdutos Menos Vendidos (Loja 2):")
display(vendas_por_produto_loja2.tail())

print("\nProdutos Mais Vendidos (Loja 3):")
display(vendas_por_produto_loja3.head())

print("\nProdutos Menos Vendidos (Loja 3):")
display(vendas_por_produto_loja3.tail())

print("\nProdutos Mais Vendidos (Loja 4):")
display(vendas_por_produto_loja4.head())

print("\nProdutos Menos Vendidos (Loja 4):")
display(vendas_por_produto_loja4.tail())
```

# 5. Frete Médio por Loja


```python
media_frete_loja1 = loja['Frete'].mean()
media_frete_loja2 = loja2['Frete'].mean()
media_frete_loja3 = loja3['Frete'].mean()
media_frete_loja4 = loja4['Frete'].mean()

print(f"Média de Frete da Loja 1: R$ {media_frete_loja1:.2f}")
print(f"Média de Frete da Loja 2: R$ {media_frete_loja2:.2f}")
print(f"Média de Frete da Loja 3: R$ {media_frete_loja3:.2f}")
print(f"Média de Frete da Loja 4: R$ {media_frete_loja4:.2f}")
```

```python
import matplotlib.pyplot as plt

# Dados
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
medias_frete = [34.69, 33.62, 33.07, 31.28]

# Gráfico de linha
plt.figure(figsize=(8, 5))
plt.plot(lojas, medias_frete, marker='o', linestyle='-', color='purple')

# Detalhes visuais
plt.title('Média de Frete por Loja')
plt.xlabel('Lojas')
plt.ylabel('Média do Frete (R$)')
plt.ylim(30, 36)
plt.grid(True, linestyle='--', alpha=0.6)

# Adiciona os valores sobre os pontos
for i, valor in enumerate(medias_frete):
    plt.text(lojas[i], valor + 0.2, f'R$ {valor:.2f}', ha='center')

plt.tight_layout()
plt.show()
```
